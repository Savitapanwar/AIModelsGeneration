```python
import numpy as np
import pandas as pd
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data= pd.read_csv("/content/fraud_transactions_synthetic.csv")
```

```python
data.head()
```

```python
x=data.drop(columns=["is_fraud"])
y=data["is_fraud"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape[1] #total of 10 neurons are supposed to be incorporated in the input layer
```

```python
from re import L
#Building the Neural Network using Tensorflow and Keras
#We will build this as a Sequential Model because there are several layers that are stacked on top of each other

model=keras.Sequential([
    layers.Input(shape=(x_train.shape[1],),name='input'), #this is input layer
    layers.Dense(32,activation='relu',name='hidden_1'), #this is the 1st hidden layer
    layers.Dense(16,activation='relu',name='hidden_2'), #this is the 2nd hidden layer
    layers.Dense(1,activation='sigmoid',name='output') #this is the output layer
])
```
